```python
def remove_exact_duplicate_rows(df: pd.DataFrame) -> pd.DataFrame:
    """Drop rows that are fully identical across all columns (keep first) and log deletions."""
    before = len(df)
    df_out = df.loc[~df.duplicated(keep="first")].copy()
    deleted = before - len(df_out)
    print(f"[dedupe] Removed {deleted} exact duplicate row(s). (before={before}, after={len(df_out)})")
    return df_out

remove_exact_duplicate_rows(df_leads)
```

```python
import numpy as np
import pandas as pd
import logging

logger = logging.getLogger(__name__)

def dedupe_and_reassign_ids(df: pd.DataFrame, id_col: str = "id") -> pd.DataFrame:
    df_out = df.loc[~df.duplicated(keep="first")].copy()

    counts = df_out[id_col].value_counts(dropna=False)
    num_duplicated_ids = int((counts > 1).sum())
    logger.info("Number of duplicated IDs: %s", num_duplicated_ids)
    print("Number of duplicated IDs: %s", num_duplicated_ids)

    dup_mask = df_out.duplicated(subset=[id_col], keep="first")
    n_to_change = int(dup_mask.sum())

    if n_to_change == 0:
        logger.info("No IDs needed reassignment.")
        print("No IDs needed reassignment.")
        return df_out

    max_id = int(df_out[id_col].max())
    new_ids = np.arange(max_id + 1, max_id + 1 + n_to_change)

    changed = df_out.loc[dup_mask, [id_col]].copy()
    changed["new_id"] = new_ids
    changed = changed.rename(columns={id_col: "old_id"})
    changed.insert(0, "row_index", changed.index)

    logger.info("Rows with ID reassigned: %s", len(changed))
    print("Rows with ID reassigned: %s", len(changed))
    for row_index, old_id, new_id in changed[["row_index", "old_id", "new_id"]].itertuples(index=False):
        logger.info("row_index=%s old_id=%s -> new_id=%s", row_index, old_id, new_id)
        print("row_index=%s old_id=%s -> new_id=%s", row_index, old_id, new_id)

    df_out.loc[dup_mask, id_col] = new_ids
    return df_out

dedupe_and_reassign_ids(df_leads, "id").head()
```
